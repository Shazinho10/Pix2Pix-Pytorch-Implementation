```python
src = "train_A"
trg = "train_B"
```

```python
import os
import cv2
import numpy as np
for file in os.listdir(src):
  full_src = os.path.join(src, file)
  full_trg = os.path.join(trg, file)

  kernel_size = (10, 10)
  image = cv2.imread(full_src)
  image = cv2.blur(image, kernel_size)
  cv2.imwrite(full_trg, image)
  print(f"dont with {file}")
```
